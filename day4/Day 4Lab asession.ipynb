###### String Formats for Float Precision

```python
x = 23.457413902458498
s = format(x, '.5f')
print(s)
```

```python
s=format(x, '.2f')
print(s)
print(x)
```

```python
x = 2.876543
y = 16.3591
print('Anjali y: {:.2f}'.format(y))
print('x long: {:.5f}, x short: {:.3f}, y: {:.2f}.'.format(x, x, y))
```

###### Example

Update a list: pop, remove, append, extend

```python
a_list = [1, 2, 3] # commas to seperate elements
print(a_list)

print("Pop %i out of a_list"%a_list.pop())
# pop the value of an index
print(a_list)
```

```python
print("Remove the string good from b_list:")
b_list = [20, True, "good1", 3,"good",2, "good"] 
print(b_list)

b_list.remove(20)
# remove a specific value (the first one in the list)
print(b_list)
```

```python
b_list.remove('good')
print(b_list)
```

```python
a_list.append(10)
print("After appending a new value, a_list is now: %s"%(str(a_list)))
```

```python
a_list.extend(b_list)
print("Merging a_list and b_list: %s"%(str(a_list)))
```

# Tuple (A special case of list whose elements cannot be changed)<br>
Initialize a tuple with paranthesis. The major difference between list and tuple is that you can alter list but not tuple.

Tuples: consists of an ordered collection of objects 
<br>Some operations similar to lists
<br>Tuples enclosed between parentheses()
<br>Immutable
<br>once created cannot be modified

```python
employee_details=('P001','John',35,40000)
print(employee_details)
```

###### Accessing componenets of a tuple
<br>To access componenets, use single slicing operator[]
<br>Syntax: tuple_name[index]

```python
print(employee_details[-1],employee_details[-4])
```

```python

print(employee_details[3])
print(employee_details[-2])
print(employee_details[-5])
# gives error
```

###### Slicing: used to access a set of elements from a tuple by creating a range of index numbers[x:y]

<br>x-index number is where the slice starts (inclusive)
<br>x: inclusive
<br>y: exclusive

```python
print(employee_details)
print(employee_details[:6])
```

```python
print(employee_details[1:3])
print(employee_details[0:3]) 
print(employee_details[:3]) 
```

###### length of a tuple
<br>len()- returns the length of a tuple
<br>Sytax: len(tuple_name)

```python
employee_details=('P001','John',35,40000)
print(employee_details)

print(len(employee_details))
```

###### Finding minimum & maximum from tuple
<br>min()- returns the lowest value in the tuple
<br>max()-r_nameeturns the highest secured by students in English
<br>Syntax: min(tuple)

```python
english=(56,85,75,96,12)
print(min(english))
print(max(english))
```

###### Combining two tuples
<br>Two tuples can be concatenated as follows: 
<br>(tuple1)+(tuple2)

```python
employee_details2=('M.Com', 'Accounts')
print(employee_details+employee_details2)
```
